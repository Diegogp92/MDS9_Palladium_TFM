```python
# Vamos a crear 3 a partir de su split

columnas_fechas = ['LLEGADA', 'SALIDA', 'FECHA_TOMA', 'FECHA_MOD', 'FECHA_CANCELACION']

    # Año
for col in columnas_fechas:
    df_Reservas_Limpio[f'{col}_ANO'] = df_Reservas_Limpio[col].astype(str).str[0:4]

    # Mes
for col in columnas_fechas:
    df_Reservas_Limpio[f'{col}_MES'] = df_Reservas_Limpio[col].astype(str).str[5:7]

    # Día del mes
for col in columnas_fechas:
    df_Reservas_Limpio[f'{col}_DIAm'] = df_Reservas_Limpio[col].astype(str).str[8:10]

    # Día de la semana
for col in columnas_fechas:
    df_Reservas_Limpio[f'{col}_DIAs'] = df_Reservas_Limpio[col].dt.weekday + 1 # +1 porque es más claro 1-7 que 0-6, así no nos liamos

# En la fase de entrenamiento ya decidiremos con cual nos quedamos
```

```python
# Estado de avance en el año

for col in columnas_fechas:
    def calcular_avance(fecha):
        if pd.isna(fecha):
            return np.nan
        dias_transcurridos = fecha.timetuple().tm_yday - 1
        dias_totales = (pd.Timestamp(year=fecha.year, month=12, day=31)).timetuple().tm_yday
        return dias_transcurridos / dias_totales

    df_Reservas_Limpio[f'{col}_AVANCE'] = df_Reservas_Limpio[col].apply(calcular_avance)
```
